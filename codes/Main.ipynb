```python
# Boone University dataset is used

# importing all the required libraries

import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mp  
import csv 

import pandas as pd 
from pandas import ExcelWriter as ew;
from pandas import ExcelFile as ef;
import math
import pywt
import nolds
from scipy.fftpack import fft, ifft
```

```python
# visualising the data

a = np.array(pd.read_excel('data.xlsx'))
# remove the first col(serial number col)

a = np.delete(arr = a, obj = 0, axis = 1) 
# axis = 0 is for row and axis = 1 for col
print(a)

m, n = np.shape(a)
print('\n \n \n ',m, n)
print('\n',a.dtype)

# n = number of subjects, m = repetition on a subject
# each row represents trial on all the 100 people
```

```python
# eeg signal extracted from an individual

for i in range(5):
    plt.plot(np.arange(0, m), a.T[i], linewidth = 0.5 )
    plt.show()
    print(len(a.T[i]), '\n')
    
```

```python
coeff =[]
for i in range(100):
#     print(i)
    x = pywt.wavedec(a.T[i], 'db4' , level = 4)
    coeff.append(x)
#     print(len(coeff[i]))
```

```python
for i in range(2):
    print("for" , i+1 , "iteration\n")
    for j in range(5):
        plt.plot(coeff[i][j], linewidth = 0.5 )
        plt.show()
    
```
