```python
a = np.array([[1, 2], [3, 4]])
```

```python
import pywt 
cA, cB = pywt.dwt2(a, 'db1')
print(cA,'\n',cB)
```

```python
a = np.array([1, 2,3 ,4, 5, 6])
cL1, cL2, cH2 = pywt.dwt(a, 'db1', level = 2)
```

```python
a = np.array([1, 2,3 ,4, 5, 6])
cL1, cL2, cH2 = pywt.dwt(a, 'db1', level = 2)
cL2, cH2 = pywt.dwt(a, 'haar')
print("cl1, ch1, cl2, ch2", '\n', cL1, '\n', cH1, '\n', cL2, '\n', cH2, '\n' )

plt.figure(1)
plt.plot(a)
plt.show()

plt.figure(2)
plt.plot(cL1,'r',cH1, 'g')
plt.show()

plt.figure(3)
plt.plot(cL2, 'r', cH2, 'g')
plt.show()
```
